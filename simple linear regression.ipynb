```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv(r"Salary_Data.csv")
df
```

```python
y = df["Salary"]
x1 = df["YearsExperience"]
```

```python
plt.scatter(x1,y)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show
```

```python
x = sm.add_constant(x1)
result = sm.OLS(y,x).fit()
result.summary()
```

```python
plt.scatter(x1,y)
yhat= 2.579e+04+9449.96*x1
fig = plt.plot(x1,yhat)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show
```
